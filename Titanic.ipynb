# 📊**Visualization**

↪ **Bar Plot**

```python
plt.figure(figsize=(8,5))
sns.barplot(x='sex',y='age',data=data,hue='survived',palette='dark:olive')
plt.title("Distribution of Passenger sex by age ")
plt.xlabel("sex")
plt.ylabel("age")
plt.show()
```

↪ **Scatter  Plot**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='age',y='fare',data=data,hue='survived',palette='dark:gray')
plt.title("Age and Fare with Survival Status")
plt.xlabel("age")
plt.ylabel("fare")
plt.show()
```

↪ **Box Plot**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='class',y='fare',data=data,hue='survived',palette='dark:green')
plt.title("Distribution of class by number of fare")
plt.xlabel("class")
plt.ylabel("fare")
plt.show()
```

# 🧹 **Data Cleaning**(Finding missing value/check Repeated Data)

"Data cleaning involves removing null values, checking for and eliminating duplicate entries, renaming or standardizing values, and identifying outliers in the dataset to ensure data quality and consistency."

**It's remove the null values**

```python
data.isnull().sum()
```

**It's Check the Duplicated values**

```python
data.duplicated()
```

**It's remove the Duplicates values**

```python
data.drop_duplicates(inplace=True)
```

**After removing duplicates from a DataFrame using drop_duplicates(), it's good practice to re-check whether any duplicates still exist** 

```python
data.duplicated().sum()
```

* **Data.dtypes → Data types**

```python
data.dtypes
```

* **Data.rename({},axis=1) → Rename the column**

```python
data.rename({'who':'person_type'},axis=1)
```

```python
data.rename({'sex':'Gender'},axis=1)
```

* **To Remove the outliers for certain variables**

```python
lb = data['fare'].quantile(0.25)
up = data['fare'].quantile(0.75)
data1 = data[(data['fare']>=lb)&(data['fare']<=up)]
sns.boxplot(data1)
```

```python
lb = data['age'].quantile(0.25)
up = data['age'].quantile(0.75)
data1 = data[(data['age']>=lb)&(data['age']<=up)]
sns.boxplot(data1)
```

# 💈 **Data Visualization And Analysis**

"After completion  data cleaning, data visualization is used to analyze and display the data through various types of plots. These include bar charts, line graphs, count plots, box plots, histograms, pie charts, scatter plots, pair plots, violin plots, and heatmaps. These visualizations help in understanding data patterns, distributions, and relationships between variables."

↪ **Bar Plot**

```python
plt.figure(figsize=(8,5))
sns.barplot(x='sex',y='parch',data=data,hue='survived',palette='dark:violet')
plt.title("Distribution of Passenger parch by Gender ")
plt.xlabel("sex")
plt.ylabel("parch")
plt.show()
```

↪ **Line Plot**

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='age',y='fare',data=data,hue='survived')
plt.title("Distribution of age number of fare ")
plt.xlabel("age")
plt.ylabel("fare")
plt.show()
```

↪ **Count Plot**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='sibsp',data=data,legend=True,edgecolor='black',hue='survived',palette='pink')
plt.title("Distribution of an sibsp ")
plt.xlabel("sibsp")
plt.show()
```

↪ **Histogram Plot**

```python
plt.figure(figsize=(8,5))
sns.histplot(bins=80,data=data,palette='dark:blue')
plt.title("Distribution of age of an fare")
plt.xlabel("age")
plt.ylabel("fare")
plt.show()
```

↪ **Box Plot**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='pclass',y='age',data=data,hue='survived',palette='dark:silver')
plt.title("Distribution of age number of pclass ")
plt.xlabel("plcass")
plt.ylabel("age")
plt.show()
```

↪ **Violin Plot**

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='sex',y='deck',data=data,hue='survived',palette='pink')
plt.title("Distribution of Passenger sex by deck ")
plt.xlabel("sex")
plt.ylabel("deck")
plt.show()
```

↪ **Scattter Plot**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='age',y='fare',data=data,hue='survived',palette='dark:olive')
plt.title("Distribution of age number of fare ")
plt.xlabel("age")
plt.ylabel("fare")
plt.show()
```

↪ **Heatmap Plot**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(numeric_only=True),annot=True,cmap='Greens')
plt.title('Heatmap of Titanic Dataset')
plt.show()
```

↪ **Pie Plot**

```python
who_counts = data['who'].value_counts()
plt.pie(who_counts, labels=who_counts.index, startangle=140, colors=["#2C0831", "#ec82e7", "#796977"],autopct='%1.1f%%', shadow=True)
plt.title('Distribution of Passengers by who')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

↪ **Pair Plot**

```python
titanic = sns.load_dataset("titanic")
pair_data = titanic[['age', 'fare', 'pclass', 'sex']].dropna()
sns.pairplot(pair_data, hue='sex', palette='RdPu')
plt.suptitle('Pair Plot of Titanic Dataset by Sex', y=1.02)
plt.show()
```
